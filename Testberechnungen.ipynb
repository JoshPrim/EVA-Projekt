### Erstellen einer Karte mit inactiven Aufzügen

```python
# Erzeugen der Karte mit inactiven Aufzügen
# Inaktive Aufzüge:
inactive = daten[daten['state'] == 'INACTIVE']

# Zoom am ausgewählten Ort
geolocator = Nominatim(user_agent="specify_your_app_name_here")

stadt = 'Frankfurt'
bundesland ='Hessen'
land = 'Deutschland'

tmp = str(stadt + ', ' + bundesland + ', ' + land)
location = geolocator.geocode(tmp)

print(location.address)
#print((location.latitude, location.longitude))

x_value = location.latitude
y_value = location.longitude


m = folium.Map(location=[x_value,y_value], zoom_start=10)

for i, row in inactive.iterrows():
    if str(row['geocoordY']) == 'nan' or str(row['geocoordX']) == 'nan':
        pass
    else:
        tmp = str(row['description'])
        folium.Marker([row['geocoordY'], row['geocoordX']], 
                      popup = tmp,
                     icon=folium.Icon(color='red', icon='info-sign')).add_to(m)


m.save('map_inactive_elivators.html')
m
```
